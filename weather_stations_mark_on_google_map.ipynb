```python
import folium
import numpy as np

# Sample list of locations: (label, latitude, longitude) - corrected comment for clarity
locations = [
    ("Larimore", 38.77261115134265, -90.21262307393965),
    ("Flint", 38.85558466257108, -90.86557211383682),
    ("Gimblin", 38.71935, -90.22528),
    ("Pine",38.69662,-90.27304),
    ("Chouteau", 38.62497, -90.22531),
    ("Carondelet", 38.56051, -90.25841),
    ("Tenbrook", 38.44593,-90.34920),
    ("Dale", 38.62810, -90.33224),
    ("Ballas", 38.63159, -90.43664),
    ("Midland",38.70995,-90.39052),
    ("Florissant", 38.78956,-90.32068),
    ("Howdershell", 38.79401, -90.37208),
    ("Duchesne",38.79850, -90.49957),
    ("Olive",38.73131, -90.66130),
    ("St Peters", 38.78793,-90.63123),
    ("Winghaven",38.75859, -90.73158),
    ("Pond", 38.582, -90.65607),
    ("Six Flags", 38.51314, -90.67511),
    ("Brush Creek", 38.48244,-90.77009),
    ("Glendale", 38.59596, -90.37693),
    ("Webster",38.59615, -90.34892),
    ("Orchard Garden",38.59148, -90.49831),
    ("Rudder",38.54306, -90.45463),
    ("Castlewood", 38.55345, -90.54605),
    ("Forder", 38.49738,-90.32538),
    ("St Charles Water", 38.78798,-90.56636),
    ("Jungermann", 38.76972,-90.58302),
    ("Wentzville",38.81168,-90.85633),
    ("May Rd", 38.81723, -90.88284)
]

# Create a base map centered roughly in the middle of the points
avg_lat = sum(lat for _, lat, _ in locations) / len(locations)
avg_lon = sum(lon for _, _, lon in locations) / len(locations)
m = folium.Map(location=[avg_lat, avg_lon], zoom_start=4, tiles="OpenStreetMap")

# Add markers for each location
for label, lat, lon in locations:
    folium.Marker(
        location=[lat, lon],
        popup=label,
        tooltip=label,
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(m)

# Add latitude/longitude grid lines
# Folium doesn't have built-in grid lines, so we simulate them with PolyLines
lat_lines = np.arange(38.5, 39, 0.1)  # every 0.1 degrees latitude
lon_lines = np.arange(-91, -90, 0.1)  # every 0.1 degrees longitude, corrected range

for lat in lat_lines:
    folium.PolyLine([(lat, -180), (lat, 180)], color="gray", weight=0.5, opacity=0.5).add_to(m)
for lon in lon_lines:
    folium.PolyLine([(-90, lon), (90, lon)], color="gray", weight=0.5, opacity=0.5).add_to(m)

# Save map to HTML file
m.save("multiple_locations_map.html")
print("Map saved as multiple_locations_map.html — open it in your browser.")
```

```python
import folium

# List of locations around St. Louis, MO (latitude, longitude, label)
locations = [
    (38.6270, -90.1994, "St. Louis Downtown"),
    (38.6350, -90.2530, "Forest Park"),
    (38.6100, -90.2700, "The Hill"),
    (38.6610, -90.2980, "Delmar Loop"),
    (38.5920, -90.3370, "Grant's Farm")
]

# Create a map centered on St. Louis
m = folium.Map(location=[38.6270, -90.1994], zoom_start=12, tiles="OpenStreetMap")

# Add markers for each location
for lat, lon, name in locations:
    folium.Marker(
        location=[lat, lon],
        popup=name,
        tooltip=name,
        icon=folium.Icon(color="blue", icon="info-sign")
    ).add_to(m)

# Add a tile layer that shows latitude/longitude grid lines
# "OpenStreetMap" doesn't show them by default, so we use "Esri.WorldImagery" + "OpenStreetMap" overlay
folium.TileLayer('Stamen Terrain', attr='Stamen Terrain').add_to(m)  # Terrain layer with some grid lines
folium.LayerControl().add_to(m)

# Save map to HTML file
m.save("st_louis_map.html")

print("Map created: st_louis_map.html (open in your browser)")
```
